### Matplotlib

```python
import matplotlib.pyplot as pl
x=[1,2,3,4]
y=[0,9,8,7]
pl.plot(x,y)
pl.show()
```

```python
import numpy as np
x=np.array([25,20,79,60])
x=np.array([23,66,49,89])
pl.plot(x,y)
pl.show()
```

```python
pl.plot(x,y,color='y')
pl.show()
```

```python
c=np.array([1,2,3,4])
pl.plot(c,c**2,color='m')
d=np.array([5,8,9,2])
pl.plot(d,d**2,color='g')
pl.show()
```

## Box Plot
   - ALso known as whisker
   - Parameters
   - minimum(1-5%)
   - First quirtile(25%)
   - Second quirtile 25%
   - max
    

```python
data=np.random.normal(50,100,150)
print(data)
```

```python
data1=np.random.randint(50,100,150)
print(data1)
data2=np.random.randint(50,78,90)
data3=np.random.randint(34,70,10)
dx=[data1,data2,data3]
dy=['a','b','c']
pl.boxplot(dx,dy)
pl.plot()
pl.savefig('boxplot.png')
```

###  PIE chart
  - Dividing each piece of the data(rows,columns)into seperate and visualize
  

```python
bc=[120,180,60,120,120]
branches=['CSE','ECE','EEE','CIV','MECH']
colors=['Cyan','Blue','Green','Purple','Yellow']
pl.pie(bc,labels=branches,startangle=90,shadow=True,explode=(0.1,0.2,0.3,0.1,0.2),autopct='%1.2f')
pl.show()
pl.savefig('piebranch.png')
```
